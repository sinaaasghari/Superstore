```python
finalpred = model_tree.predict(df_ff)
finalpred_prob= model_tree.predict_proba(df_ff)

finalpred
```

```python
finalpred_prob
```

```python
def column(matrix, i):
    return [row[i] for row in matrix]

column(finalpred_prob, 1)
```

```python
output = dp.copy()
output['Prediction - whitch mode Numb'] = finalpred
output['Prediction - Probilty to First Class'] = column(finalpred_prob,0 )
output['Prediction - Probilty to Same Day'] =  column(finalpred_prob, 1)
output['Prediction - Probilty to Second Class'] =  column(finalpred_prob,2 )
output['Prediction - Probilty to Standard'] = column(finalpred_prob, 3)
output['Prediction - whitch mode Name'] = 'Empty'
output['Prediction - whitch mode Name'][output['Prediction - whitch mode Numb'] == 0] = 'First Class'
output['Prediction - whitch mode Name'][output['Prediction - whitch mode Numb'] == 1] = 'Same Day'
output['Prediction - whitch mode Name'][output['Prediction - whitch mode Numb'] == 2] = 'Second Class'
output['Prediction - whitch mode Name'][output['Prediction - whitch mode Numb'] == 3] = 'Standard'

```
